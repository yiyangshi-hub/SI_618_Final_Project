```python
from census import Census
from us import states

c = Census("62c6704b10bcb447c467e027fe446144c20d7ccb")

# B01001_001E represents the total population
# B19013_001E: median income
# B25077_001E: median house value
# B02001_002E: # of white 
# B02001_003E: # Black/African American
# B02001_004E: American Indian, Alaska Native
# B02001_005E: Asian
# B02001_006E: Native Hawaiian, other Pacific Islanders
# B02001_007E: Some other race alone
variables = ('B01001_001E','B19013_001E','B25077_001E',
             'B02001_002E','B02001_003E','B02001_004E',
             'B02001_005E','B02001_006E','B02001_007E')  

# Retrieve data for Illinois
CT_data = pd.DataFrame(c.acs5.state_county(variables, states.CT.fips, '*'))
CO_data = pd.DataFrame(c.acs5.state_county(variables, states.CO.fips, '*'))
SC_data = pd.DataFrame(c.acs5.state_county(variables, states.SC.fips, '*'))
WA_data = pd.DataFrame(c.acs5.state_county(variables, states.WA.fips, '*'))
WI_data = pd.DataFrame(c.acs5.state_county(variables, states.WI.fips, '*'))
AZ_data = pd.DataFrame(c.acs5.state_county(variables, states.AZ.fips, '*'))
TX_data = pd.DataFrame(c.acs5.state_county(variables, states.TX.fips, '*'))

```

```python
states_census = pd.concat([CT_data, CO_data, SC_data, WA_data, WI_data, AZ_data, TX_data], axis=0)
```

```python
WI_data
```

```python
states_census = states_census.rename(columns={
    'B01001_001E' : 'population',
    'B19013_001E' : 'median_income',
    'B25077_001E' : 'median_house_val',
    'B02001_002E' : 'num_of_white',
    'B02001_003E' : 'num_of_black',
    'B02001_004E' : 'num_of_indigenous',
    'B02001_005E' : 'num_of_asian',
    'B02001_006E' : 'num_of_islanders',
    'B02001_007E' : 'num_of_others'
})

state_names = {'09' : 'CT', '08':'CO', '45':'SC', 
               '53':'WA', '55':'WI', '04':'AZ', '48':'TX'}

states_census['state'] = states_census['state'].map(state_names)
```
